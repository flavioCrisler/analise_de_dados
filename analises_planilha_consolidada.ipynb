## Análise de margem de lucro

```python
df.head()
```

### Agrupando dados
#### Por seguimento

```python
df.groupby('segmento').sum().round(2) #.mean()| .median()
```

```python
df.groupby('segmento')['valor_total'].sum()
```

```python
colunas = ['valor_total_desconto', 'custo_total', 'lucro']
df_segmento = df.groupby('segmento')[colunas].sum().round(2)
```

```python
df_segmento['margem_lucro'] = (df_segmento['lucro'] / df_segmento['valor_total_desconto'] * 100)
```

```python
df_segmento.round(2)
```

#### Por produto

```python
colunas = ['valor_total_desconto', 'custo_total', 'lucro']
df_produto = df.groupby('produto')[colunas].sum().round(2)
df_produto['margem_lucro'] = (df_produto['lucro'] / df_produto['valor_total_desconto'] * 100)
df_produto.round(2)
```

#### Por país


```python
colunas = ['valor_total_desconto', 'custo_total', 'lucro']
df_pais = df.groupby('pais')[colunas].sum().round(2)
df_pais['margem_lucro'] = (df_pais['lucro'] / df_pais['valor_total_desconto'] * 100)
df_pais.round(2)
```
